# Build team_features.csv for match outcome modeling
This notebook loads StatsBomb event CSV files, computes team-level features,adds a simple match outcome label per team, and saves the result as`data/team_features.csv` for the scikit-learn training script.

```python
from pathlib import Pathimport pandas as pdimport sysproject_src = Path.cwd() / "src"if str(project_src) not in sys.path:    sys.path.append(str(project_src))from football_ai.preprocessing.feature_engineering import compute_team_featuresDATA_DIR = Path("data")
```

## Load one matchFor now we will use a single match events CSV. You can extend this to loopover multiple files later.

```python
events_path = DATA_DIR / "raw" / "comp_43_season_3" / "events_7525.csv"events = pd.read_csv(events_path)events.head()
```

## Compute team-level features

```python
team_features = compute_team_features(events)team_features
```

## Add match metadata and outcome labelsYou need to define the outcome (result) for each team in the match basedon the scoreline. Here we hard-code the result for match 7525 as anexample. Replace with the real labels as needed.

```python
match_id = 7525  # update if you use a different matchteam_features = team_features.reset_index().rename(columns={team_features.index.name or "index": "team"})team_features["match_id"] = match_id# Example labels: first row win, second row loss.# Replace this list with the actual outcomes for each team.team_features["result"] = ["win", "loss"]team_features
```

## Save to data/team_features.csv

```python
out_path = DATA_DIR / "team_features.csv"out_path.parent.mkdir(parents=True, exist_ok=True)team_features.to_csv(out_path, index=False)out_path.resolve()
```
